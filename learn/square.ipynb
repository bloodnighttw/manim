```python
from manim import *
import numpy as np
```

```python
%%manim --format=gif -qm -v WARNING CircleToSquare

class CircleToSquare(Scene):
    def construct(self):
        circle = Circle()  # create a circle
        circle.set_color(WHITE)

        square = Square()  # create a square
        square.rotate(PI / 4)  # rotate a certain amount

        self.play(Create(circle))  # animate the creation of the square
        self.play(Transform(circle, square))  # interpolate the square into the circle
        self.play(FadeOut(square))  # fade out animation
```

```python
%%manim --format=gif -qm -v WARNING CircleToSquare2

class CircleToSquare2(Scene):
    def construct(self):
        circle1 = Circle()  # create a circle
        circle1.set_color(WHITE)
        
        circle2 = Circle()
        circle2.set_color(WHITE)
        circle2.next_to(circle1, RIGHT)

        square = Square()  # create a square
        square.rotate(PI / 4)  # rotate a certain amount

        self.play(Create(circle1),Create(circle2))  # animate the creation of the square
        self.play(FadeOut(circle1),FadeOut(circle2))  # fade out animation
```

# Using .animate

```python
%%manim --format=gif -qm -v WARNING CircleToSquare3

class CircleToSquare3(Scene):
    def construct(self):
        circle1 = Circle()  # create a circle
        circle1.set_color(WHITE)

        circle2 = Circle()
        circle2.set_color(BLUE)
        circle2.next_to(circle1, RIGHT)
        
        self.add_foreground_mobjects(circle2,circle1) #

        self.play(Create(circle1))  # animate the creation of the square
        self.play(circle2.animate.scale(1.5).next_to(circle1,np.array([-1,0,0],dtype=float)))
        self.play(circle2.animate.scale(0).rotate(PI / 4).fade(0), FadeOut(circle1))

```
